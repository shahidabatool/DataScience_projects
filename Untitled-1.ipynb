```python
df_plot=df[pd.to_numeric(df['latitude'],errors='coerce').notnull()& 
           pd.to_numeric(df['longitude'],errors='coerce').notnull()]

df_plot['latitude'] = df_plot['latitude'].astype(float)
df_plot['longitude'] = df_plot['longitude'].astype(float)
df_sample = df_plot.sample(5000, random_state=42)
```

```python
import folium 

world_map=folium.Map(location=[20,0],zoom_start=2)
for idx, row in df_sample.iterrows():
    region = row['region_txt']
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=2,
        color='red',
        fill=True,
        fill_opacity=0.5
    ).add_to(world_map)

# -------------------------------
# 4. Save map to HTML
# -------------------------------
world_map.save("terrorism_map.html")
print("Map created successfully! Open 'terrorism_map.html' to view it.")
```
